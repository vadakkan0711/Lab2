```python
# importing libraries
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns
%matplotlib inline
from ucimlrepo import fetch_ucirepo
```

```python
# Fetch dataset
breast_cancer_wisconsin_diagnostic = fetch_ucirepo(id=17)

X = breast_cancer_wisconsin_diagnostic.data.features 
y = breast_cancer_wisconsin_diagnostic.data.targets["Diagnosis"] 

# Verify shapes before concatenation
print(f"Shape of X: {X.shape}")  # Features
print(f"Shape of y: {y.shape}")  # Target
df = pd.concat([X, y], axis=1)

# Display first few rows
df.head()
```

```python
plt.figure(figsize=(8, 7))
sns.countplot(x="Diagnosis", data=df, palette='magma')  # Use "Diagnosis" (capital D)
plt.title("Distribution of Diagnosis")
plt.show()
```

```python
# mapping categorical values to numerical values
df['Diagnosis']=df['Diagnosis'].map({'B':0,'M':1})
```

```python
from sklearn.model_selection import train_test_split

# Splitting data 
X_train, X_test, y_train, y_test = train_test_split(
                df.drop('Diagnosis', axis=1),  
                df['Diagnosis'],  
                test_size=0.2,
                random_state=42)

# Print shapes of the datasets
print("Shape of training set:", X_train.shape)
print("Shape of test set:", X_test.shape)
```

```python
from sklearn.preprocessing import StandardScaler

# Initialize StandardScaler
ss = StandardScaler()

# Fit and transform only on the training set
X_train = ss.fit_transform(X_train)

# Transform (not fit) the test set using the same scaler
X_test = ss.transform(X_test)
```

## Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

# Initialize Random Forest 
rfc = RandomForestClassifier(n_estimators=300, random_state=42, n_jobs=-1)

# Train the model
rfc.fit(X_train, y_train)

# Predict on test data
predictions4 = rfc.predict(X_test)

# Calculate accuracy
rfc_acc = accuracy_score(y_test, predictions4)

# Print accuracy
print(f"Accuracy of the Random Forest Model: {rfc_acc:.4f}")

# Classification Report
print("\nClassification Report:\n", classification_report(y_test, predictions4))

# Confusion Matrix
print("\nConfusion Matrix:\n", confusion_matrix(y_test, predictions4))
```
